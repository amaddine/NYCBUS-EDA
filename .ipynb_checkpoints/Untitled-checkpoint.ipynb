```python
import pandas as pd 
import numpy as np 
```

```python
data_frame = pd.read_csv('clean.csv')
```

```python
data_frame.head()
```

```python
def clean_up(x):
    if x == 'Yes':
        x = True
    elif x == 'No':
        x = False
        
```

```python
data_frame['Has_Contractor_Notified_Schools_bool'] = data_frame['Has_Contractor_Notified_Schools'].map(lambda x: clean_up(x))
data_frame['Has_Contractor_Notified_Parents_bool'] = data_frame['Has_Contractor_Notified_Parents'].map(lambda x: clean_up(x))
```
